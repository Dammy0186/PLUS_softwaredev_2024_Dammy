```python
import folium
import pandas as pd
```

```python
#working on how to get the Galleries Exhibition space open studios,create a simple map, and add points to each of the location using folium in python
#Getting the locations and saving to a data frame
#Pandas allows us to import data from various file formats such as CSV,JSON etc
#A connection to the csv file is made and accessed with pandas
```

```python
location = "https://data.smartdublin.ie/dataset/3f9c4989-2a63-4b6a-aa6f-54d205181d55/resource/1baddead-e218-41c7-8dfd-d58d5d768007/download/030124-dataset-galleries-exhibition-spaces-open-studios.csv"
```

```python
galleries_exhibition_space_open_studios = pd.read_csv(location)
```

```python
galleries_exhibition_space_open_studios
```

```python
The galleries_exhibition_space_open_studios_locations has been imported a total of 24 data
Now i need to keep the latitude and longitude and the name(Title) of the locations. 
The first two columns will allow me to map the location and the last two will be use to give each location pin a name
```

```python
galleries_exhibition_space_open_studios = galleries_exhibition_space_open_studios[["Latitude", "Longitude", "Title","DccAreaNew"]]
```

```python
#The next step is to create a map of the location using the location parameters.
#I zoomed to 5 then i later zoom to 15
```

```python
map = folium.Map(location=[galleries_exhibition_space_open_studios.Latitude.mean(), galleries_exhibition_space_open_studios.Longitude.mean()], zoom_start=15, control_scale=True)
```

```python
map
```

```python
#Now i am showing the location for each galleries on the map
```

```python
for index, location_info in galleries_exhibition_space_open_studios.iterrows():
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], popup=location_info["Title"]).add_to(map)
```

```python
map
```

```python
#These are the galleries_exhibition_space_open_studio location in dublin, but we have them in a single color
# Rather than having a single colour representing the Folium markers, we can change the colour based on DccAreaNew within our data
```

```python
DccAreaNew_colour = {'Central':'red', 'South_East':'green', 'South_Central':'blue', 'North_Central':'brown'}

map = folium.Map(location=[galleries_exhibition_space_open_studios.Latitude.mean(), galleries_exhibition_space_open_studios.Longitude.mean()], 
                 zoom_start=15, control_scale=True)
#Loop through each row in the dataframe
for i,row in galleries_exhibition_space_open_studios.iterrows():
    #Setup the content of the popup
    iframe = folium.IFrame(f'Title: {str(row["Title"])} \n DccAreaNew: {str(row["DccAreaNew"])}')
    
    #Initialise the popup using the iframe
    popup = folium.Popup(iframe, min_width=300, max_width=300)
    
    try:
        icon_color = DccAreaNew_colour[row['DccAreaNew']]
    except:
        #Catch nans
        icon_color = 'gray'
    
    #Add each row to the map
    folium.Marker(location=[row['Latitude'],row['Longitude']],
                  popup = popup, 
                  icon=folium.Icon(color=icon_color, icon='')).add_to(map)
```
